# Activation functions: sigmoid, tanh, ReLU


One of the problems of traditional activation functions such a sigmoid or a tanh is that it flattens quickly (value going to 0/1 for sigmoid or -1/1 for tanh). This is can cause difficulty with backpropagation where partial derivatives can go to zero quickly as well, as a result the weigths cannot be updated and the model cannot learn.

ReLU functions help to achieve fast convergence, so the model trains quickly. It is also faster to compute derivatives on essentially linear function. Allows each neuron to express a stronger opinion! 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from pathlib import Path
Path('figs/').mkdir( exist_ok=True)
```

### Sigmoid

$\sigma(x)=\frac{1}{1+e^{−x}}$

```python
x = np.linspace(-10, 10, 1000)
sigma = 1 / (1 + np.exp(-x) )

plt.figure(figsize=(10, 5))
plt.plot(x, sigma)
plt.legend(['sigmoid function'])
plt.show()
```

### tanh

$\tanh(x)=\frac{e^x−e^{−x}}{e^x+e^{−x}}$

```python
x = np.linspace(-10, 10, 1000)
tanh = ( 2 / (1 + np.exp(-2*x) ) ) -1

plt.figure(figsize=(10, 5))
plt.plot(x, tanh)
plt.legend(['tanh'])
plt.show()
```

### ReLU or Rectified Linear Unit

$f(x)=\max(0,x)$

```python
x = np.linspace(-10, 10, 1000)
relu = np.maximum(0, x)

plt.figure(figsize=(10, 5))
plt.plot(x, relu)
plt.legend(['Relu'])
plt.show()
```

### Sigmoid+tanh

```python
plt.figure(figsize=(10, 5))
plt.plot(x, sigma, label='sigmoid')
plt.plot(x, tanh, label='tanh')
plt.ylim(-1.5,1.5)
plt.legend()
plt.savefig('figs/sigmoid_tanh.png', dpi=300)
plt.show()
```

### Combined

```python
plt.figure(figsize=(10, 5))
plt.plot(x, sigma, label='sigmoid')
plt.plot(x, tanh, label='tanh')
plt.plot(x, relu, label='ReLU')
plt.ylim(-1.5,1.5)
plt.legend()
plt.savefig('figs/combined.png', dpi=300)
plt.show()
```
